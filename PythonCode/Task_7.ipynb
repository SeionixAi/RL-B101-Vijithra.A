**Write a Python program to check whether a string contains only digits.**

```python
def is_only_digits(text):
    # Step 1: Check each character
    for char in text:
        if not char.isdigit():
            return False
    return f"contains only digits"
```

```python
print(is_only_digits("123"))
```

```python
print(is_only_digits("abc"))
```

```python
print(is_only_digits("134#$%"))
```

**Write a Python program to find the longest word in a string.**

```python
def find_longest_word(text):
    words = text.split()
    longest_word = ""

    for word in words:
        # If the current word is longer than our previous longest, update it
        if len(word) > len(longest_word):
            longest_word = word

    return longest_word
```

```python
print(find_longest_word("I am a Data Science Enthusiast"))
```

```python
print(find_longest_word("apple"))
```

**Write a Python program to capitalize the first and last character of each word in a string.**

```python
def capitalize_ends(text):
    words = text.split()
    result = []

    for word in words:
        if len(word) > 1:
            # Capitalize first, keep middle, capitalize last
            new_word = word[0].upper() + word[1:-1] + word[-1].upper()
            result.append(new_word)
        else:
            result.append(word.upper())


    return " ".join(result)
```

```python
print(capitalize_ends("hello world"))
```

```python
print(capitalize_ends("python"))
```

**Draw a right-angled triangle pattern using stars.**

```python
def right_triangle(n):
    for i in range(1, n + 1):
        print("* " * i)
```

```python
print(right_triangle(5))
```

**Draw an inverted right-angled triangle pattern.**

```python
def inverted_triangle(n):
    for i in range(n, 0, -1):
        print("* " * i)
```

```python
print(inverted_triangle(10))
```
