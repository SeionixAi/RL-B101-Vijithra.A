To count the number of vowels in a given string


```python
def count_vowels(text):
    vowels = "aeiou"
    count = 0
    for char in text.lower():
        if char in vowels:
            count += 1
    return count

```

```python
print(count_vowels("Data Science"))
```

```python
print(count_vowels("Syllogism"))
```

To check whether a given string is a Palindrome


```python
def is_palindrome(text):
  lower_text=text.lower()
  reversed_text=lower_text[::-1] # Reverse the lowercase text
  if lower_text==reversed_text:
     return True
  else:
    return False
```

```python
re1=is_palindrome("Madam")
print(re1)
```

```python
re2=is_palindrome("Syllogism")
print(re2)
```

To check whether two strings are Anagrams

```python
def is_anagram(str1,str2):
   string1=str1.lower()
   string2=str2.lower()
   sorted_str1=sorted(string1)
   sorted_str2=sorted(string2)
   if sorted_str1==sorted_str2:
     return True
   else:
     return False
```

```python
Eg1= is_anagram("listen","Silent")
print(Eg1)
```

```python
Eg2= is_anagram("Malayalam","malayalam")
print(Eg2)
```

```python
Eg3= is_anagram("Artificial ","Intelligence")
print(Eg3)
```

Common & BasicSeaborn Uses

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load example dataset
df = sns.load_dataset("titanic")
```

```python
# Histogram (distplot/deprecated â†’ use displot)
sns.histplot(df['age'].dropna())
plt.title("Histogram of Age")
plt.show()
```

```python
#Count plot (for categorical data)
sns.countplot(x='class', data=df)
plt.title("Passenger Class Count")
plt.show()
```

```python
#Box plot
sns.boxplot(x='class', y='age', data=df)
plt.title("Age Distribution by Class")
plt.show()
```

```python
# Violin plot
sns.violinplot(x='class', y='age', data=df)
plt.title("Violin Plot - Age by Class")
plt.show()
```

```python
# Bar plot (summary stats)
sns.barplot(x='sex', y='survived', data=df)
plt.title("Survival Rate by Sex")
plt.show()


```

```python
#Scatter plot (relplot or scatterplot)
sns.scatterplot(x='age', y='fare', data=df)
plt.title("Age vs Fare")
plt.show()
```

```python
#Pair plot
sns.pairplot(df[['age', 'fare', 'pclass', 'survived']].dropna())
plt.suptitle("Pair Plot", y=1.02)
plt.show()

```

```python
#Swarm plot
sns.swarmplot(x='class', y='age', data=df)
plt.title("Swarm Plot - Age by Class")
plt.show()

```

```python
#Strip plot
sns.stripplot(x='class', y='age', data=df, jitter=True)
plt.title("Strip Plot - Age by Class")
plt.show()
```
