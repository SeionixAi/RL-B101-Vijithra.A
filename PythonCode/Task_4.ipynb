**Data Preprocessing**

```python
# check missing values
df.isna().sum()
```

```python
df.dtypes
```

```python
# Checking duplicate values
df.duplicated().sum()
```

```python
df.duplicated()
```

```python
df=df.drop_duplicates()
```

```python
# converting data types
df['Order Date']=pd.to_datetime(df['Order Date'])
df['Ship Date']=pd.to_datetime(df['Ship Date'])
```

```python
df.info()
```

```python
# convert to lowercase
df['Product Name']=df['Product Name'].str.lower()
```

```python
df['Customer Name']=df['Customer Name'].str.lower()
```

```python
# Rename column names
df.rename(columns={'Customer Name':'Customer_Name'},inplace=True)
```

```python
df.columns
```

```python
# calculating profit margin
df['Profit_Margin']=df['Profit']/df['Sales']
```

```python
df['Profit_Margin']
```

**Data** **Visualization**

Histograms are used to visualize the distribution of sales values and understand how frequently different sales ranges occur.

```python
# Histogram
plt.hist(df['Sales'])
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.title('Histogram of Sales')
plt.show()
```

```python
#Distribution plot
plt.figure(figsize=(8,5))
sns.histplot(df['Sales'], bins=30, kde=True)
plt.title("Sales Distribution")
plt.show()
```

Count plots are used to show the frequency of categorical variables such as customer segments and product categories, helping to identify dominant groups.


```python
# count plot
plt.figure(figsize=(10,5))
sns.countplot(x='Category', data=df)
plt.title("Product Category Count")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='Segment', data=df)
plt.title("Customer Segment Count")
plt.show()
```

A correlation heatmap is used to identify relationships between numerical variables such as sales, profit, and discount. This helps understand how variables influence each other.

```python
#Heatmap
plt.figure(figsize=(10,6))
sns.heatmap(df.select_dtypes(include='number').corr(),
            annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

### To analyze monthly sales trends, a new feature order_month was created by extracting the month from the order_date column.


```python
# Extract year and month from order date
df['order_year'] = df['Order Date'].dt.year
df['order_month'] = df['Order Date'].dt.month
```

```python
# Time based
monthly_sales = df.groupby('order_month')['Sales'].sum()

plt.figure(figsize=(8,5))
monthly_sales.plot()
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.show()
```

```python
# category wise profit
plt.figure(figsize=(8,5))
sns.barplot(x='Category', y='Profit', data=df)
plt.title("Profit by Product Category")
plt.xlabel("Category")
plt.ylabel("Total Profit")
plt.show()
```

```python
# saving the cleaned dataset
df.to_csv("cleaned_sample_superstore.csv", index=False)
```
