  ## Write a Python program to count the number of consonants in a given string.


```python
def count_consonants(text):
    vowels = "aeiou"
    count = 0
    for char in text.lower():
        if char.isalpha() and char not in vowels:
            count += 1
    return count
```

```python
print(count_consonants("Python Programming"))
```

```python
print(count_consonants("Data Analytics"))
```

```python
print(count_consonants("Artificial Intelligence"))
```

  ## Write a Python program to find the first non-repeating character in a string.

```python
def first_non_repeating(text):  
    lower_text = text.lower()    # Convert the text to lowercase
    char_count = {}              # Count the frequency of each character
    for char in lower_text:
        if char in char_count:
            char_count[char] += 1
        else:
            char_count[char] = 1
    for char in lower_text:         # Step 3: Find the first character with count 1
        if char_count[char] == 1:
            return char
    
    return None
```

```python
print(first_non_repeating("Malayalam"))
```

```python
print(first_non_repeating("Data Science"))
```

  ## Write a Python program to remove all duplicate characters from a string.

```python
def remove_duplicates(text):
    result = ""
    for char in text:                        
        if char not in result:   # If the character is not already in our result, add it
            result += char
    return result
```

```python
print(remove_duplicates("Malayalam"))
```

```python
print(remove_duplicates("character"))
```
