**Find the factorial of a number**

```python
def factorial_of(num):
  result=1
  for i in range(1,num+1):
    result*=i
  return result
```

```python
print(factorial_of(25))
```

```python
print(factorial_of(3))
```

**Find the Largest and Smallest number in a list**


```python
def find_min_max(num):
  if not num:
    return None
  largest=num[0]
  smallest=num[0]

  for numbers in num:
    if numbers < smallest:
            smallest = numbers
    if numbers > largest:
            largest = numbers

  return smallest, largest
```

```python
print(find_min_max([1,2,3,4,0]))
```

```python
print(find_min_max([12,-1,0,3]))
```

**Fibbonacci Sequence**

```python
def fibbonacci_sequence(n):
    a, b = 0, 1
    for _ in range(n):
        print(a, end=" ")
        a, b = b, a + b
        a = b
        b = a+b
```

```python
fibbonacci_sequence(10)
```

**Create a Square Frame Pattern**

```python
def square_frame(n):
    for i in range(n):
        for j in range(n):
            if i == 0 or i == n-1 or j == 0 or j == n-1:
                print("*", end=" ")
            else:
                print(" ", end=" ")
        print()
```

```python
square_frame(10)
```

```python
square_frame(100)
```

**Sum of Digits in a Number**

```python
def sum_of_digits(n):
    total = 0
    for digit in str(n):
        total += int(digit)
    return total
```

```python
sum_of_digits(3987)
```
